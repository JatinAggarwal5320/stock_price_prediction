```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
start = '2005-01-01'
end ='2025-01-01'
stock = 'GOOG'

data =yf.download(stock, start, end)
```

```python
ema_100_days = data.Close.ewm(span=100, adjust=False).mean()
ema_200_days = data.Close.ewm(span=200, adjust=False).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Close Price', color='blue', linewidth=2)
plt.plot(data.index, ema_100_days, label='100-day EMA', color='orange', linestyle='dashed')
plt.plot(data.index, ema_200_days, label='200-day EMA', color='red', linestyle='dashed')
plt.title(f'{stock} Stock Price with 100-day & 200-day EMA', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Stock Price', fontsize=12)
plt.legend()
plt.grid(True)

plt.show()
```

```python
data.dropna(inplace=True)
```

```python
data_train= pd.DataFrame(data.Close[0: int(len(data)*0.80)])
data_test= pd.DataFrame(data.Close[int(len(data)*0.80): len(data)])
```

```python
data_train.shape[0]
```

```python
data_test.shape[0]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_train_scale = scaler.fit_transform(data_train)
```

```python
x= []
y=[]
for i in range(100 , data_train_scale.shape[0]):
    x.append(data_train_scale[i-100:i])
    y.append(data_train_scale[i,0])
    
               
```

```python
x,y = np.array(x), np.array(y)
```

```python
from keras.layers import Dense , Dropout , LSTM
from keras.models import Sequential
```

```python
model = Sequential()
model.add(LSTM(units = 50 , activation = 'relu' , return_sequences = True,
               input_shape = ((x.shape[1],1))))
model.add(Dropout(0.2))

model.add(LSTM(units = 60 , activation = 'relu' , return_sequences = True))
model.add(Dropout(0.3))

model.add(LSTM(units = 80 , activation = 'relu' , return_sequences = True))
model.add(Dropout(0.4))

model.add(LSTM(units = 120 , activation = 'relu'))
model.add(Dropout(0.4))

model.add(Dense(units = 1))
```

```python
model.compile(optimizer = 'adam' , loss = 'mean_squared_error')
```

```python
model.fit(x, y, epochs =50 ,batch_size = 32 , verbose=1)
```

```python
past_100_days = data_train.tail(100)
```

```python
data_test = pd.concat([past_100_days, data_test], ignore_index=True)
```

```python
data_test_scale = scaler.fit_transform(data_test)
```

```python
x= []
y=[]
for i in range(100 , data_test_scale.shape[0]):
    x.append(data_test_scale[i-100:i])
    y.append(data_test_scale[i,0])
    
x,y = np.array(x) , np.array(y)
```

```python
y_predict = model.predict(x)
```

```python
scale = 1/scaler.scale_
```

```python
y_predict = y_predict* scale
```

```python
y = y* scale
```

```python
plt.figure(figsize=(10, 8))
plt.plot(y, label="Actual Stock Price", color='blue', linewidth=2)
plt.plot(y_predict, label="Predicted Stock Price", color='red', linestyle='dashed')
plt.title(f'{stock} Stock Price Prediction', fontsize=14)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Stock Price', fontsize=12)
plt.legend()
plt.show()
```

```python
model.save('Stock Price Prediction Model.keras')
```
